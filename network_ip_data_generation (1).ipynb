```python
import random
import time

def generate_network_traffic():
    while True:
        # Simulate a source IP address
        source_ip = ".".join(str(random.randint(0, 255)) for _ in range(4))

        # Simulate a destination IP address
        dest_ip = ".".join(str(random.randint(0, 255)) for _ in range(4))

        # Simulate a timestamp
        timestamp = int(time.time())

        # Simulate the amount of data transferred
        data_transferred = random.randint(1, 1000)

        # Print or process the generated network traffic data
        print(f"Source IP: {source_ip} | Destination IP: {dest_ip} | Timestamp: {timestamp} | Data Transferred: {data_transferred} bytes")

        # Sleep for a random duration between 0.1 to 2 seconds
        time.sleep(random.uniform(0.1, 2))

# Generate network traffic data
generate_network_traffic()
```
